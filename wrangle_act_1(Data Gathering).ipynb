```python
# reading the api_key, api_key_secret, 
#access_token and access_token_secret from a json file.

with open('config_file.json', 'r') as file:
   config_data = json.load(file)

api_key = config_data['api_key']
api_key_secret = config_data['api_key_secret']
access_token = config_data['access_token'] 
access_token_secret = config_data['access_token_secret']
```

```python
# accesing the twitter api with the config_data items.
auth =tw.OAuthHandler(api_key,api_key_secret)
auth.set_access_token(access_token,access_token_secret)
api= tw.API(auth,wait_on_rate_limit=True)
api
```

```python
# extracting required field for analysis from the tweet_j list
# creating a list of keys
key_list=[]
for key in tweet_j[0]._json:
    x=key
    key_list.append(x)
print(key_list)  
```

```python
#create a list of dictionaries containing content of status
created_at=[]
id_=[]
full_text=[]
entities=[]
source=[]
in_reply_to_status_id=[]
in_reply_to_user_id=[]
in_reply_to_screen_name=[]
user=[]
geo=[]
coordinates=[]
place=[]
contributors=[]
is_quote_status=[]
retweet_count=[]
favorite_count=[]
favorited=[]
retweeted=[]
lang=[]
for status in tweet_j:
    content=status._json
    created_at.append(content['created_at'])
    id_.append(content['id'])
    full_text.append(content['full_text'])
    entities.append(content['entities'])
    source.append(content['source'])
    in_reply_to_status_id.append(content['in_reply_to_status_id'])
    in_reply_to_user_id.append(content['in_reply_to_user_id'])
    in_reply_to_screen_name.append(content['in_reply_to_screen_name'])
    user.append(content['user'])
    geo.append(content['geo'])
    coordinates.append(content['coordinates'])
    place.append(content['place'])
    contributors.append(content['contributors'])
    is_quote_status.append(content['is_quote_status'])
    retweet_count.append(content['retweet_count'])
    favorite_count.append(content['favorite_count'])
    favorited.append(content['favorited'])
    retweeted.append(content['retweeted'])
    lang.append(content['lang'])
    

tweet_dict={'created_at':created_at,'id':id_,'full_text':full_text,'entities':entities,'source':source,\
            'in_reply_to_status_id':in_reply_to_status_id,'in_reply_to_user_id':in_reply_to_user_id,\
           'in_reply_to_screen_name':in_reply_to_screen_name,'user':user,'geo':geo,'coordinates':coordinates,\
            'place':place,'contributors':contributors,'is_quote_status':is_quote_status,'retweet_count':retweet_count,\
           'favorite_count':favorite_count,'favorited':favorited,'retweeted':retweeted,'lang':lang}
```

```python
# create a DataFrame of all tweet status
df=pd.DataFrame(tweet_dict)
df.head()
```
