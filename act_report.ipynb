Look at these beautiful faces they are super stars, the best in their classes. They  are so good, they won competitions without even knowing it. Always at their best yet always being themselves. In this blogpost today, they will be recognised for their achievements.

Thanks to their humans for capturing such amazing moments and posting it on weratedog twitter page. Weratedog is a twitter page that allow people to post such amazing moments with their dog while rating them and sharing some interesting comments about them. in return, the tweet will be liked and retweeted depending on the level of interest and response it is able to pull. Weratedog has over 4 million followers and has earned several international recognition. Therefore, to be best in any class here is not a mean feat.

This report is generated from the result of an analysis which was carried out in partial fulfilment of the requirement to earn a Udacity nanodegree in data analytics. It involved collecting archived tweet data of the weratedog twitter account, and requesting even more information from the twitter api using the tweepy library inorder to develop robust insight from the analysis. 

A third source of data employed for this analysis is a neural network image classification of the dog images by breeds. This was already done for us by the curator of the project,but as this project was particulaly designed to test my data wrangling skills, several effort still had to be made to get it into a usable form inorder to generate these amazing insights I am about to share with you.

These insights can be broadly divided into two categories. 
- The first category showcase the broad pattern that can be observed across the entire data set.
- The category two is about our super stars who were best in their classes.


#### Most common dog breed
Across the dataset it can be observed that there are about **108** different breeds of dog submitted on the weratedog archive. The most common among them is the **golden retriever**. since the population of the weratedog following is a sample of the population of dog lovers. we can conclude that the **golden retriever** is the most prefered dog breed among dog lovers.

#### Most common/popular dog growth stage

Since we extracted the dog_growth_stage information from the text accompanying the tweet post, most of the dog_growth stage information were not available and were set to **none** on the dog_growth_stage field. ignoring those for which their growth stage information was not available, the analysis shows that the most common dog growth stage is the **pupper**. This can be observed by considering that the pupper dog growth stage in the **fig 2** has more data points in the swarm plot.

Also, the mean most retweeted, dog_growth_stage is the **doggo**. This mean value might have been greatly affected by the outlier in the plot. Therefore a better estimation of the class retweet value will be the median. Then the most retweeted dog_growth_stage by median, is the **pupper**.This information can be seen in the **fig 3**. if you wish to know the difference between each of these dog_growth_stage, then  [here is a very a short but interesting video](https://www.youtube.com/watch?v=ah6fmNEtXFI&list=PLUoebdZqEHTygG_DML0kNfbjnkv-MLnwi&index=1)

#### Sentiment analysis on the tweet text

Using **Roberta**, a pretrained Natural Language Processing model by **hugging face**, sentiment analysis was also carried out on the tweet text to classify them into three categories, ``negative``, ``neutral``, ``positive`` based on the mood of the text. With this classification, the audience response can be observed. It was observed that tweets with **positive** mood got the most retweets and tweets with **neutral** mood got the least retweet as can be seen from the graph in **fig 4**.

The graph 4 shows the retweet_count for the top 5 tweets in the negative, neutral and positive sentiment category. 

###### fig 4 retweet count of the best in class of sentiment category.

The tweet with the most retweet is a positive tweet with tweet_id **678399652199309312**. It has  retweet_count value of **28,631** in the year **2015** where the mean retweet_count was **912.711048**.

The tweet text reads thus, `This made my day. 12/10 please enjoy`.
